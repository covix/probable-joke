```python
import pandas as pd

%matplotlib notebook
import matplotlib.pyplot as plt

%matplotlib notebook
%matplotlib notebook
```

```python
df = pd.read_csv('/Users/covix/Projects/probable-joke/scripts/num_frames_loud.txt', sep='\s',
                 names=['signer', 'sign', 'light', 'count'], engine='python')
```

```python
new_df = pd.DataFrame(columns=range(0, 14))

for i in range(1, 44):
    odd, even = [], []
    tmp = df.ix[df.sign == i, ['count']]
    
    for t in tmp.itertuples():
        if t.Index % 2 == 0:
            even.append(t.count)
        else:
            odd.append(t.count)
    
    new_df = new_df.append([odd, even])    


new_df.columns = range(1, 15)
new_df.index = range(1, len(new_df) + 1)
new_df = new_df.astype(pd.np.int)
```

```python
plt.figure()
new_df.boxplot()
plt.title("Boxplot for signers")
```

```python
new_df = pd.DataFrame(columns=range(0, 43))

for i in range(1, 15):
    odd, even = [], []
    tmp = df.ix[df.signer == i, ['count']]
    
    for t in tmp.itertuples():
        if t.Index % 2 == 0:
            even.append(t.count)
        else:
            odd.append(t.count)
    
    new_df = new_df.append([odd, even])    


new_df.columns = range(1, 44)
new_df.index = range(1, len(new_df) + 1)
new_df = new_df.astype(pd.np.int)
```

```python
plt.figure()
new_df.boxplot()
plt.title("Boxplot for classes")
```

```python
plt.figure()
plt.suptitle("Length distribution for class")
for i in range(1, 44):
    plt.subplot(11, 4, i)
    df_s = df.ix[df.sign == i, ['count']]
    df_s.index = range(1, 29)
    plt.plot(df_s, 'o-', label=i, alpha=0.3)
    plt.legend(numpoints=1)
```

```python
plt.figure()
plt.suptitle("Length distribution per signer")
for i in range(1, 15):
    plt.subplot(3, 5, i)
    df_s = df.ix[df.signer == i, ['count']]
    df_s.index = range(1, 87)
    plt.plot(df_s, 'o-', label=i, alpha=0.3)
    plt.legend(numpoints=1)
```
